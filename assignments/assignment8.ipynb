```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
```

```python
alpha = np.linspace(0, 1, 10000)
f = 1 - np.power(alpha, 2) + np.divide(4 * np.power(alpha, 2), 1 + alpha) - np.divide(4 * alpha, 1 + alpha)
```

```python
plt.plot(alpha, f)
plt.xlabel(r'$\alpha$')
plt.ylabel(r'$f(\alpha)$')
plt.grid()
plt.savefig('../images/oving8-oppg3.png')
plt.show()
```
